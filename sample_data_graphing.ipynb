```python
# Please see previous editor's note regarding CPU-memory related inquiries



# provides various subplots given the dataframe, the names of subplots, graphing types, x axis, and number of 
# rows and columns
def plot_subplots(dataframe_with_name, dataframe_with_data, subplot_names, x_axis, nrows, ncols):
    # requires type(data_frame_with_name) == type(data_frame_with_data) == dataframe object
    assert(isinstance(subplot_names, list) or isinstance(subplot_names, tuple))
    assert(isinstance(x_axis, str))
    assert(isinstance(nrows, int) and isinstance(ncols, int))
    for name in subplot_names:
        if name not in list(dataframe_with_name["name"]):
            print("Desired population \'"+name+"\' Not in list of names in the dataframe. Exiting function...")
            return
    if x_axis not in list(dataframe_with_data.columns):
        print('Cannot find x-axis variable \"'+x_axis+'\" in given dataframe. Exiting function...')
        return
    elif (nrows*ncols != len(subplot_names)):
        print ("Improper dimensions of nrows and ncols relative to the length of subplot_names. Exiting function...")
        return
    subset_dataframe = dataframe_with_name[dataframe_with_name["name"].isin(subplot_names)] 
    # assert the number of trials per plot is the same
    num_colors_per_frame = int(len(list(subset_dataframe["column"])))
    # change the cmap name if you don't like the colors
    # cmap name reference: https://matplotlib.org/stable/tutorials/colors/colormaps.html
    cmap = plt.cm.get_cmap("viridis", num_colors_per_frame)
    color_i = 0
    fig, axes = plt.subplots(nrows, ncols)
    subplot_name_i = 0;
    for row in range(nrows):
        for col in range(ncols):
            current_name = subset_dataframe["name"].iloc[subplot_name_i]
            while(subplot_name_i < len(subset_dataframe["name"]) and subset_dataframe["name"].iloc[subplot_name_i] == current_name):
                y_axis = subset_dataframe["column"].iloc[subplot_name_i]
                if y_axis not in list(dataframe_with_data.columns):
                    print("Cannot find y-axis variable \'"+y_axis+"\' in given dataframe with data. Exiting function...")
                    return
                if (y_axis != x_axis):
                    if (nrows != 1 and ncols != 1):
                        axes[row, col].plot(dataframe_with_data[x_axis],dataframe_with_data[y_axis], color=cmap(color_i))
                    elif(nrows != 1):
                        assert(ncols == 1)
                        axes[row].plot(dataframe_with_data[x_axis],dataframe_with_data[y_axis], color=cmap(color_i))
                    else:
                        axes[col].plot(dataframe_with_data[x_axis],dataframe_with_data[y_axis], color=cmap(color_i))
                        axes[col].set_xlabel(x_axis)

                color_i = color_i + 1
                subplot_name_i = subplot_name_i + 1
            color_i = 0
            if (nrows != 1 and ncols != 1):
                axes[row, col].set_ylabel(current_name)
                axes[row, col].set_xlabel(x_axis)
            elif(nrows != 1):
                axes[row].set_ylabel(current_name)
                axes[row].set_xlabel(x_axis)
            else:
                axes[col].set_ylabel(current_name)
                axes[col].set_xlabel(x_axis)
    plt.show()
    plt.clf()
    
    
    
# plots a list of desired multiple action channels against each other
def plot_multiple_action_channels(figure, dataframe_with_name, dataframe_with_data, x_axis, name):
    # requires type(data_frame_with_name) == type(data_frame_with_data) == dataframe object
    assert(isinstance(x_axis, str) and isinstance(name, str))
    if name not in list(dataframe_with_name["name"]):
        print("Desired population \'"+name+"\' Not in list of names in the dataframe. Exiting function...")
        return
    elif x_axis not in list(dataframe_with_data.columns):
        print('Cannot find x-axis variable \"'+x_axis+'\" in given dataframe. Exiting function...')
        return
    figure.add_subplot(1,1,1)
    subset_dataframe = dataframe_with_name[dataframe_with_name["name"] == name]    
    # change the cmap name if you don't like the colors
    # cmap name reference: https://matplotlib.org/stable/tutorials/colors/colormaps.html
    cmap = plt.cm.get_cmap("viridis", len(list(subset_dataframe["column"])))
    color_i = 0
    
    for column in subset_dataframe["column"]:
        if column not in list(dataframe_with_data.columns):
            print("Cannot find y-axis variable \'"+column+"\' in given dataframe with data. Exiting function...")
            return
        if (column != x_axis):
            plt.plot(dataframe_with_data[x_axis],dataframe_with_data[column], color=cmap(color_i))
        color_i = color_i + 1
    plt.xlabel(x_axis)
    plt.legend(subset_dataframe["column"])
    plt.title('Plot of '+name+' Action Channels over '+x_axis+":")
    plt.show()
    plt.clf()
    
    
    
# plots a list of desired multiple action channels against each other
def plot_multiple_action_channels_snippet(figure, dataframe_with_name, dataframe_with_data, x_axis, name, axes_dims):
    ## requires axes[0] == x0, axes[1] == x1, axes[2] == y0, axes[3] == y1
    ## requires type(data_frame_with_name) == type(data_frame_with_data) == dataframe object
    assert(isinstance(x_axis, str) and isinstance(name, str))
    assert(isinstance(axes_dims, list) or isinstance(axes_dims, tuple))
    assert(axes_dims[0] < axes_dims[1] and axes_dims[2] < axes_dims[3])
    if name not in list(dataframe_with_name["name"]):
        print("Desired population \'"+name+"\' Not in list of names in the dataframe. Exiting function...")
        return
    elif x_axis not in list(dataframe_with_data.columns):
        print('Cannot find x-axis variable \"'+x_axis+'\" in given dataframe. Exiting function...')
        return
    figure.add_subplot(1,1,1)
    subset_dataframe = dataframe_with_name[dataframe_with_name["name"] == name]    
   
    # change the cmap name if you don't like the colors
    # cmap name reference: https://matplotlib.org/stable/tutorials/colors/colormaps.html
    cmap = plt.cm.get_cmap("viridis", len(list(subset_dataframe["column"])))
    color_i = 0
    
    for column in subset_dataframe["column"]:
        if column not in list(dataframe_with_data.columns):
            print("Cannot find y-axis variable \'"+column+"\' in given dataframe with data. Exiting function...")
            return
        if (column != x_axis):
            plt.plot(dataframe_with_data[x_axis],dataframe_with_data[column], color=cmap(color_i))
        color_i = color_i + 1
    plt.xlabel(x_axis)
    plt.legend(subset_dataframe["column"])
    plt.xlim(axes_dims[0], axes_dims[1])
    plt.ylim(axes_dims[2], axes_dims[3])
    plt.title('Plot of '+name+' Action Channels over '+x_axis+":")
    plt.show()
    plt.clf()

    
    
def plot_trials(figure, dataframe, x_axis, y_axis, trials):
    ## requires dataframe_with_x_axis and dataframe_with_y_axis are dataframe objects
    ## requires the total number of trials is equal to the number of rows in the dataframe
    assert(isinstance(x_axis, str) and isinstance(y_axis, str))
    assert(isinstance(trials, int))
    
    if x_axis not in list(dataframe.columns):
        print('Cannot find x-axis variable \"'+x_axis+'\" in given dataframe. Exiting function...')
        return
    elif (trials < 0 or trials > dataframe.shape[0]):
        print('Number of trials \"'+str(trials)+'\" is either negative or bigger than the amount of total trials in dataframe. Exiting function...')
        return
    elif y_axis not in list(dataframe.columns):
        print('Cannot find y-axis variable \"'+y_axis+'\" in given dataframe. Exiting function...')
        return
    elif (x_axis == y_axis):
        print("Improper format: x_axis and y_axis are equal. Exiting function...")
        return
    subset_dataframe = dataframe.iloc[0:trials]
    plot_scatterplot(figure, subset_dataframe, x_axis, y_axis)
    

# Issues:

# Margin Stuff

# Vertical Lines for Highlighting Key Features

# "Automating" Bounds

# Plotting Other Plotting Types In Subplots besides lineplots

# Merging Different Plots From Different Dataframes

# Joining Different Plots Together

# Plotting API


```
